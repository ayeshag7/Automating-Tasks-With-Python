```python
# Importing the required packages.
from captcha.image import ImageCaptcha
```

```python
# Specifying the image size.
image = ImageCaptcha(width=400, height=200)

# Entering the text.
text = input("Enter captcha: ")

# Generating the image.
captcha = image.generate(text)

# Writing the image onto a file and saving it.
image.write(text, "E:\\CAPTCHA.png")

# Displaying the image using pillow.
from PIL import Image
Image.open("E:\\CAPTCHA.png")
```
